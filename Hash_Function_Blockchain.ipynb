```python
import hashlib
```

```python
def hash_txt(text, algorithm="sha256"):
  hash_obj=hashlib.new(algorithm)
  hash_obj.update(text.encode())
  return hash_obj.hexdigest()
```

```python
message = input("Enter the sentence : ")
hash_value= hash_txt(message)
print(f"Your Input Sentence : {message}")
print(f"Hashed Sentence : {hash_value}")
```
